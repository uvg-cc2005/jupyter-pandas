## Hojas de estilo (Style Sheets)

Matplotlib tiene [hojas de estilo](http://matplotlib.org/gallery.html#style_sheets) que pueden ser utilizados para que sus gráficas se vean más bonitas. Estas hojas de estilo incluyen plot_bmh, plot_fivethirtyeight, plot_ggplot y más. Estas básicamente crean un conjunto de reglas de estilo que obecen sus gráficas. Se recomienda usarlas ya que hacen que todas sus gráficas tengan la misma apariencia y se vean más profesionales. Es posible crear nuestras propias hojas de estilo si quisiéramos que las gráficas de nuestra empresa, organización, club, etc., todas tuvieran la misma apariencia (es algo tedioso crear una, sin embargo).

Esta es la forma de usarlas.

**Sin plt.style.use() sus gráficas se ven así:**

```python
df1['A'].hist()
```

Invocar el estilo:

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Ahora sus gráficas se ven así:

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

# Tipos de gráficas

Pandas tiene varios tipos de gráficas integradas, la mayoría son gráficas estadísticas, por naturaleza:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

También se pueden invocar de esta forma df.plot(kind='hist') o reemplazar el argumento de tipo con cualquiera  de los tipos en la lista anterior (e.g. 'box','barh', etc..)
___

Hagamos una inspección de estas!

## Area

```python
df2.plot.area(alpha=0.4)   # alpha es un parametro que indica la cantidad de traslucidez
```

## Gráficas de barras (Barplots)

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histogramas

```python
df1['A'].plot.hist(bins=50)
```

## Gráficas de línea (Line Plots)

```python
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

## Gráficas de dispersión (Scatter Plots)

```python
df1.plot.scatter(x='A',y='B')
```

Se puede usar c para colorer basado en el valor de otra columna
Utilice cmap para indicar el mapeo de colores (colormap) a utilizar. 
Para ver todos los mapas de colores, vea: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

O utilice s para indicar que el tamaño es proporcional al valor en otra columna. El parámetro s debe ser un arreglo, no solo el nombre de una columna:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

## Gráficas de caja (BoxPlots)

```python
df2.plot.box() # Tambien se puede pasar un argumento para agrupar by= 
```

## Gráfica de compartimentos hexagonales (Hexagonal Bin Plot)

Utiles para datos Bivariados, es una alternativa a las gráficas de dispersión:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

Es todo!  Si han trabajado antes directamente con matplotlib, verán que este método de graficar es bastante más fácil.  Es un balance entre facilidad de uso y el control sobre la figura. Muchas de las llamadas para graficar también aceptan argumentos adicionales que están disponibles en las llamadas directas a matplotlib.
